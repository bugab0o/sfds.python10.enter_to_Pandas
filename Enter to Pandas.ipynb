DATAFRAME КАК СТРУКТУРА ДАННЫХ

Самый простой способ создания DataFrame — из словаря, 
ключами которого являются имена столбцов будущей таблицы, 
а значениями — списки, в которых хранится содержимое этих столбцов:

```python
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'square': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})
countries_df.index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ']
display(countries_df)
```

Также DataFrame можно создать из вложенного списка, 
внутренние списки которого будут являться строками новой таблицы:

```python
countries_df = pd.DataFrame(
    data = [
        ['Англия', 56.29, 133396],
        ['Канада', 38.05, 9984670],
        ['США', 322.28, 9826630],
        ['Россия', 146.24, 17125191],
        ['Украина', 45.5, 603628],
        ['Беларусь', 9.5, 207600],
        ['Казахстан', 17.04, 2724902]
    ],
    columns= ['country', 'population', 'square'],
    index = ['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ']
)
countries_df.loc[['UA', 'CA', 'RU'], ['square', 'population']]

# display(countries_df)
```

```python
import numpy as np
import pandas as pd

solar_data = pd.read_csv(r'./solarpower.csv')
solar_data.replace(-1, np.nan, inplace=True)

solar_data['cum_power'].interpolate(method='linear', inplace=True)

# display(solar_data.head(5))
# display(solar_data['cum_power'].mean())

# solar_data.info()
solar_data['day_power'] = solar_data['cum_power'] - solar_data['cum_power'].shift(1)
solar_data['date'] = pd.to_datetime(solar_data['date'])
solar_data['month_year'] = solar_data['date'].dt.to_period('M')
monthly_average = solar_data.groupby('month_year')['day_power'].mean()
average_march_2017 = int(monthly_average['2017-03'].round())

display(average_march_2017)

monthly_average['day_power'].plot(
    kind='line',
    lw=3, 
    title='Monthly average',
    xlabel='Year - month',
    ylabel='Average',
    figsize=(10, 5)
);

```
